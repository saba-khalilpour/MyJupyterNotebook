Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
pip install requests beautifulsoup4 pandas
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for Tesla's financials on Yahoo Finance
url = 'https://finance.yahoo.com/quote/TSLA/financials'

# Send a request to fetch the HTML content of the page
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table that contains the revenue data (Income Statement table)
table = soup.find_all('table')[0]  # Index 0 for the first table (Income Statement)

# Extract the headers (columns) from the table
headers = [header.get_text() for header in table.find_all('th')]

# Extract the rows (financial data)
rows = table.find_all('tr')[1:]  # Skipping the header row

# Initialize an empty list to store revenue data
revenue_data = []

# Loop through each row to extract the revenue data (usually in the first or second column)
for row in rows:
    columns = row.find_all('td')
    if len(columns) > 1:
        # Assuming revenue is in the second column
        revenue_data.append(columns[1].get_text())

# Clean up the revenue data and create a DataFrame
revenue_df = pd.DataFrame(revenue_data, columns=['Revenue'])

# Display the result
print(revenue_df)
```

Question 5: Plot Tesla Stock Graph



```python
pip install matplotlib yfinance
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to plot the stock graph
def make_graph(data):
    # Plot the 'Close' price of Tesla stock
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    
    # Add a title to the graph
    plt.title('Tesla Stock Price (TSLA)', fontsize=16)
    
    # Label the x and y axes
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    
    # Show the plot with a grid
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2025-01-01')

# Reset the index to make 'Date' a column
tesla_data.reset_index(inplace=True)

# Call the function to plot the data
make_graph(tesla_data)
```
